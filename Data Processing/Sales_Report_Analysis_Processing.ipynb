```python
import pandas as pd
```

```python
url = 'https://docs.google.com/spreadsheets/d/1VKeq1Txmb5HmYbaQxDkEfl-ZvDtX4HxF/edit?usp=sharing&ouid=104525316223540898596&rtpof=true&sd=true'
path = 'https://drive.google.com/uc?export=download&id=' + url.split('/')[-2]
```

```python
regions = pd.read_excel(path, 'Regions')
```

```python
regions
```

```python
regions = regions.assign(Countries = regions.Countries.str.split('; ')).explode('Countries').reset_index(drop=True).rename({'Countries': 'Country'}, axis=1)
regions['Country'] = regions['Country'].str.strip()
regions['Region'] = regions['Region'].str.strip()
regions = regions[['Country', 'Region']]
regions = regions.reset_index()
regions
```

```python
orders = pd.read_excel(path, 'Orders')
orders
```

```python
orders.info()
```

```python
orders['Total Due'] = orders['Units Sold'] * orders['Price']
orders['Total Cost'] = orders['Units Sold'] * orders['Cost']
orders['Gross Profit'] = orders['Total Due'] - orders['Total Cost']
orders['Lead Time (day)'] = orders['Ship Date'] - orders['Order Date']
orders['Lead Time (day)'] = orders['Lead Time (day)'].dt.days
orders['priority_name'] = orders['Order Priority'].map({'L': 'Low', 'M': 'Medium', 'H': 'High', 'C': 'Critical'})
orders['priority_num'] = orders['priority_name'].map({'Low': 4, 'Medium': 3, 'High': 2, 'Critical': 1})
orders = orders[['Order ID', 'Order Priority', 'Order Date', 'Ship Date', 'Lead Time (day)', 'Item Type', 'Units Sold', 'Price', 'Cost', 'Total Due', 'Total Cost', 'Gross Profit', 'Sales Channel', 'Country', 'priority_name', 'priority_num']]
orders
```

```python
def create_date_table(start='2000-01-01', end='2050-12-31'):
  df = pd.DataFrame({"Date": pd.date_range(start, end)})
  df["Year"] = df.Date.dt.year
  df["Month Name"] = df.Date.dt.month_name()
  df["Day Name"] = df.Date.dt.day_name()
  df['Month number'] = df.Date.dt.month
  df['Day of week'] = df.Date.dt.dayofweek
  return df

dim_date = create_date_table(start=orders['Order Date'].min(), end=orders['Ship Date'].max())
dim_date
```

```python
dim_date.to_csv('dim_date.csv', index=False, encoding='utf-8')
orders.to_csv('orders.csv', index=False, encoding='utf-8')
regions.to_csv('regions.csv', index=False, encoding='utf-8-sig')
```
